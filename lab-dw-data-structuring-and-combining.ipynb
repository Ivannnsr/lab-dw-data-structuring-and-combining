```python
import pandas as pd
import numpy as np
```

```python
url_file1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url_file2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url_file3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

file1 = pd.read_csv(url_file1)
file2 = pd.read_csv(url_file2)
file3 = pd.read_csv(url_file3)
```

```python
# Convertir los nombres de las columnas a minÃºsculas
file1.columns = file1.columns.str.lower()
file2.columns = file2.columns.str.lower()
file3.columns = file3.columns.str.lower()
```

```python

# Reemplazar espacios en blanco por guiones bajos
file1.columns = file1.columns.str.replace(' ', '_', regex=True)
file2.columns = file2.columns.str.replace(' ', '_', regex=True)
file3.columns = file3.columns.str.replace(' ', '_', regex=True)
```

```python
# Reemplazar "st" por "state"
file1.rename(columns={'State': 'state'}, inplace=True)
file2.rename(columns={'State': 'state'}, inplace=True)
```

```python
print("Columnas de file1:")
print(file1.columns)

print("\nColumnas de file2:")
print(file2.columns)

print("\nColumnas de file3:")
print(file3.columns)
```

```python
column_order = [
    'customer', 'state', 'gender', 'education', 'customer_lifetime_value',
    'income', 'monthly_premium_auto', 'number_of_open_complaints',
    'policy_type', 'vehicle_class', 'total_claim_amount'
    ]

# Reordenar las columnas de file2
file1 = file1[column_order]
file2 = file2[column_order]
file3 = file3[column_order]


# Verificar el nuevo orden de las columnas
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
"""
file1_info = file1.info()
file2_info = file2.info()
file3_info = file3.info()

file1_head, file2_head, file3_head
"""
```

```python
# Combinar los tres datasets
lab_dataframe1 = pd.concat([file1, file2, file3], ignore_index=True)

# Verificar el resultado
print("Dimensiones del dataset combinado:", lab_dataframe1.shape)
print("Columnas del dataset combinado:", lab_dataframe1.columns)

# Mostrar las primeras filas del dataset combinado
print(lab_dataframe1.head())
```

```python
# Imprimir los valores Ãºnicos de las columnas relevantes
columns = ['customer', 'state', 'gender', 'education', 
           'customer_lifetime_value', 'income', 
           'monthly_premium_auto', 'number_of_open_complaints', 
           'policy_type', 'vehicle_class', 'total_claim_amount']

for col in columns:
    print(f"Column: {col}, Type: {lab_dataframe1[col].dtype}")

```

```python
for column in lab_dataframe1.columns:
    print(f"Column: {column}")
    print(lab_dataframe1[column].unique())
    print("-" * 40)  # Separador entre columnas
```

```python
# Convertir todos los valores a minÃºsculas
lab_dataframe1['state'] = lab_dataframe1['state'].str.lower()
lab_dataframe1['gender'] = lab_dataframe1['gender'].str.lower()
lab_dataframe1['education'] = lab_dataframe1['education'].str.lower()
lab_dataframe1['policy_type'] = lab_dataframe1['policy_type'].str.lower()
lab_dataframe1['vehicle_class'] = lab_dataframe1['vehicle_class'].str.lower()

# Reemplazar valores especÃ­ficos
state_mapping = {
    'cali': 'california',
    'az': 'arizona',
    'wa': 'washington'
}

lab_dataframe1['state'] = lab_dataframe1['state'].replace(state_mapping)



# Reemplazar valores especÃ­ficos
gender_mapping = {
    'femal': 'f',
    'female': 'f',
    'male': 'm'
}

lab_dataframe1['gender'] = lab_dataframe1['gender'].replace(gender_mapping)

# Reemplazar 'bachelors' por 'bachelor'
lab_dataframe1['education'] = lab_dataframe1['education'].replace('bachelors', 'bachelor')

for column in lab_dataframe1.columns:
    print(f"Column: {column}")
    print(lab_dataframe1[column].unique())
    print("-" * 40)  # Separador entre columnas
```

```python
challenge2 =  "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

df2= pd.read_csv(challenge2)
```

```python
df2.columns = df2.columns.str.lower()
df2.columns = df2.columns.str.replace(' ', '_', regex=True)
```

```python
print("Columnas de df2:")
print(df2.columns)
```

```python
df2.drop('unnamed:_0', axis=1, inplace=True)
```

```python
df2
```

```python
for column in df2.columns:
    print(f"Column: {column}")
    print(df2[column].unique())
    print("-" * 40)  # Separador entre columnas
```

```python
summary_table = df2.pivot_table(
    values='total_claim_amount',  # columna que se sumarÃ¡
    index='sales_channel',  # columna de agrupaciÃ³n
    aggfunc='sum',  # sumar el total
    margins=False  # No aÃ±adir un total general
).round(2)  # Redondear a 2 decimales

# Mostrar la tabla resumen
print(summary_table)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Crear la tabla de resumen con pivot_table
clv_table = df2.pivot_table(
    values='customer_lifetime_value',  # columna que se promediarÃ¡
    index='gender',  # filas: gÃ©nero
    columns='education',  # columnas: nivel educativo
    aggfunc='mean',  # calcular promedio
    margins=False  # No aÃ±adir un total general
).round(2)  # Redondear a 2 decimales

# Mostrar la tabla resumen
print(clv_table)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Crear una tabla de resumen con formato largo
complaints_summary = (
    df2.groupby(['policy_type', 'month'])  # Agrupar por policy_type y month
    .size()  # Contar las ocurrencias
    .reset_index(name='complaints_count')  # Convertir Ã­ndices en columnas y nombrar la columna de conteo
    .sort_values(by=['policy_type', 'month'])  # Ordenar por policy_type y mes
)

# Mostrar la tabla en formato largo
print(complaints_summary)
```
